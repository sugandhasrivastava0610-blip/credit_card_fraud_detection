```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('https://www.kaggle.com/datasets/arockiaselciaa/creditcardcsv')
```

This dataset is highly unbalanced

0 -> Normal Transaction

1 -> Fraudulent transaction

```python
legit= credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()
```

Under Sampling

```python
legit_sample = legit.sample(n=492)
```

Concatinating two Dataframes

```python
new_dataset= pd.concat([legit_sample, fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting data into features & targets

```python
x= new_dataset.drop(columns='Class', axis=1)
y= new_dataset['Class']
```

```python
print(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

Model Training- Logistic Regression

```python
model_scaled = LogisticRegression()
model_scaled.fit(x_train_scaled, y_train)
```

Model Evaluation

```python
X_train_prediction = model_scaled.predict(x_train_scaled)
```

```python
train_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print('Accuracy on training data:',train_data_accuracy)
```

```python
x_test_prediction = model_scaled.predict(x_test_scaled)
```

```python
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print("Accuracy score on test data:", test_data_accuracy)
```

```python
import joblib

joblib.dump(model_scaled, "fraud_model.pkl")

print("Model saved successfully")
```

```python
import gradio as gr
import pandas as pd
import joblib

model = joblib.load("fraud_model.pkl")

def predict_fraud(*features):

    columns = [
        "V1","V2","V3","V4","V5","V6","V7","V8",
        "V9","V10","V11","V12","V13","V14","V15","V16",
        "V17","V18","V19","V20","V21","V22","V23","V24",
        "V25","V26","V27","V28","Amount"
    ]

    data = pd.DataFrame([features], columns=columns)

    prediction = model.predict(data)[0]

    if hasattr(model, "predict_proba"):
        confidence = max(model.predict_proba(data)[0])
    else:
        confidence = "N/A"

    result = (
        "🚨 Fraudulent Transaction"
        if prediction == 1
        else "✅ Legitimate Transaction"
    )

    return result, confidence


inputs = [
    gr.Number(label=f"V{i}")
    for i in range(1, 29)
]

inputs.append(
    gr.Number(label="Amount")
)

demo = gr.Interface(
    fn=predict_fraud,
    inputs=inputs,
    outputs=[
        gr.Textbox(label="Prediction"),
        gr.Textbox(label="Confidence")
    ],
    title="Credit Card Fraud Detection",
    description="Predict whether a transaction is fraudulent or legitimate."
)

demo.launch()
```
